```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split as split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
maternal_data = pd.read_csv('Maternal Health Risk Data Set.csv')
```

```python
maternal_data.head()
```

```python
# maternal_data = maternal_data.drop(maternal_data.index[maternal_data.HeartRate == 7])
# maternal_data = maternal_data.drop(["HeartRate"], axis=1)
```

```python
x=maternal_data.drop('RiskLevel',axis=1)
y=maternal_data.RiskLevel
```

```python
X_train, X_test, y_train, y_test  = train_test_split(x, y, test_size=0.2, random_state=30)
```

```python
maternal_model = RandomForestClassifier(random_state=100)
maternal_model.fit(X_train, y_train)
```

```python
maternal_model.score(X_train, y_train)
```

```python
maternal_model.score(X_test,y_test)
```

```python
y_pred = maternal_model.predict(X_test)
```

```python
filename = 'maternal_model.sav'
pickle.dump(maternal_model, open(filename, 'wb'))
```
